```python
%matplotlib inline

import sys
sys.path.insert(0,'/Library/TeX/Root/bin/x86_64-darwin/')

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('top-20.csv')
df = df.sort_values(by='word')
```

```python
ax = df.plot.bar('word', figsize=(12, 6))
ax.set(title='Top 20 words', xlabel='word', ylabel='count')
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('../Plot.png')
```
